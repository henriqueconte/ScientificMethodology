# 1. Choose a firstname and analyse its frequency along time. Compare several firstnames frequency

First, let's analyse the name Emmanuel. To do it, we can filter the dataset to get all rows with name Emmanuel, group by year and sum the number of "Emmanuel's" of each department. We will get the result below.

```r
FirstNames %>%
    filter(preusuel == "EMMANUEL") %>%
    group_by(preusuel, annais)%>%
    summarise_at(vars(nombre), list(nombre = sum)) 
```

```r
plot(FirstNames %>%
    filter(preusuel == "EMMANUEL") %>%
    group_by(annais)%>%
    summarise_at(vars(nombre), list(nombre = sum)), main = "EMMANUEL")
```

As we can see, we had most Emmanuel's in the years 1960 until 1990. That might explain why France's president is called Emmanuel! (He was born in 1977).

Let's check other names, such as Marie and Clara.

```r
plot(FirstNames %>%
    filter(preusuel == "MARIE") %>%
    group_by(annais)%>%
    summarise_at(vars(nombre), list(nombre = sum)), main = "MARIE")

plot(FirstNames %>%
    filter(preusuel == "CLARA") %>%
    group_by(annais)%>%
    summarise_at(vars(nombre), list(nombre = sum)), main = "CLARA")
```

As we can see, Marie was a pretty common name in the past, and its presence is decreasing among the years. This is probably related to religion, since Marie is a religious name.

Besides, the name Clara became very common from 1990 to 2010, and then quickly decreased. I don't know the exact reason for that, but I suppose it's because we were receiving big influence from American names and culture in this period. 

# 2. Establish by gender the most given firstname by year. Analyse the evolution of the most frequent firstname.

First, let's get the most given male firstnames by year. To do that, we first need to filter all the male names to reduce our dataset size, then we should remove noisy information like "_PRENOMS_RARES"_ and year "XXXX". These information are probably undefined values that shouldn't be in the dataset, and if we don't filter them we would only see "_PRENOMS_RARES"_ in the previous years. 

Finally, after filtering the right information, we group it by year and firstname, getting the total amount of each firstname by using the "summarise" method. Since this method will give us all the names from the year and their sum, we want to filter only the names that have the maximum frequency from each year. 

```r
frequentMaleNames = FirstNames %>% 
    filter(sexe == 1) %>%
    filter(preusuel != "_PRENOMS_RARES") %>%
    filter(annais != "XXXX") %>%
    group_by(annais, preusuel) %>% 
    summarise(frequency = sum(nombre)) %>% 
    filter(frequency == max(frequency)) 
```

### Result 

Let's see the resulting dataset we got from our operation:

```r
frequentMaleNames
```

As we can see, the most common name from 1900 until 1957 was Jean! Besides, Gabriel has been a huge hit in the past 6 years as well. Let's see a graphical representation from 1900-1915 and 2007-2021:

```r
ggplot(head(frequentMaleNames, 15), aes(annais, frequency, fill=preusuel)) + 
    geom_col() + 
    xlab("Year") + 
    ylab("Frequency")

ggplot(tail(frequentMaleNames, 15), aes(annais, frequency, fill=preusuel)) + 
    geom_col() + 
    xlab("Year") + 
    ylab("Frequency")
```

### Female results

```r
frequentFemaleNames = FirstNames %>% 
    filter(sexe == 2) %>%
    filter(preusuel != "_PRENOMS_RARES") %>%
    filter(annais != "XXXX") %>%
    group_by(annais, preusuel) %>% 
    summarise(frequency = sum(nombre)) %>% 
    filter(frequency == max(frequency)) 

frequentFemaleNames
```

For men, the name Jean was popular from 1900 to 1957. For women, we have the same phenomena but with the name Marie! I believe this happens because both names are religious. Let's check a graphical representation for women as well:

```r
ggplot(head(frequentFemaleNames, 15), aes(annais, frequency, fill=preusuel)) + 
    geom_col() + 
    xlab("Year") + 
    ylab("Frequency")

ggplot(tail(frequentFemaleNames, 15), aes(annais, frequency, fill=preusuel)) + 
    geom_col()  + 
    xlab("Year") + 
    ylab("Frequency")
```
