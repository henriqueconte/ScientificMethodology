# Getting the (wrong) picture from the data: kindergarten alternative pedagogy

## Setup the data

```r
group1 = read.csv("kindergarten_group1.csv", header = T, sep=",")
group2 = read.csv("kindergarten_group2.csv", header = T, sep=",")
```

We will try to see the improvement of Group 1 from the first year to second year, and from second year to third year. To do that, we will get the difference of values from each section and make a histogram to see the amount that most students improved.

```r
second_year_improv_group1 = group1$score.in.moyenne.section - group1$score.in.petite.section
third_year_improv_group1 = group1$score.in.grande.section - group1$score.in.moyenne.section
hist(second_year_improv_group1)
hist(third_year_improv_group1)
```

### Conclusions group 1

As we can see, students on group 1 improve a bit on the second year and improve a lot on the third year.



Now, let's do the same calculation for group 2.

```r
second_year_improv_group2 = group2$score.in.moyenne.section - group2$score.in.petite.section
third_year_improv_group2 = group2$score.in.grande.section - group2$score.in.moyenne.section
hist(second_year_improv_group2)
hist(third_year_improv_group2)
```

### Conclusions group 2

As we can see, the group 2 had better improvements on the second year, but worst improvements on the third year. In fact, the third year got even worse results in some cases when compared to the second year.

## Conclusions 

Given this dataset, the best teaching methodology is the classical pedagogy.
